```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataframe
file_path = 'df_merged.parquet'
df = pd.read_parquet(file_path)

# Información básica del dataframe
print("Información del dataframe:")
print(df.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(df.describe(include='all'))

# Verificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Distribución de las calificaciones (stars)
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], bins=20, kde=True)
plt.title('Distribución de las calificaciones')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: La mayoría de las calificaciones se encuentran entre 3 y 4.5, con un pico alrededor de 4 estrellas, lo que indica que los clientes generalmente están satisfechos con los locales.")

# Distribución de la polaridad
plt.figure(figsize=(10, 6))
sns.histplot(df['polarity'], bins=20, kde=True)
plt.title('Distribución de la polaridad')
plt.xlabel('Polaridad')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: La polaridad tiene una distribución ligeramente sesgada hacia la derecha, lo que sugiere que las reseñas tienden a ser más positivas que negativas.")

# Distribución de la subjetividad
plt.figure(figsize=(10, 6))
sns.histplot(df['subjectivity'], bins=20, kde=True)
plt.title('Distribución de la subjetividad')
plt.xlabel('Subjetividad')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: La subjetividad muestra una distribución casi uniforme, lo que indica que las reseñas tienen un balance entre subjetivas y objetivas.")

# Conteo de reviews por ciudad
plt.figure(figsize=(12, 8))
city_counts = df['city_x'].value_counts().head(20)
sns.barplot(y=city_counts.index, x=city_counts.values, palette='viridis')
plt.title('Top 20 ciudades con más reviews')
plt.xlabel('Número de reviews')
plt.ylabel('Ciudad')
plt.show()
print("Conclusión: Las ciudades con más reviews son lugares clave para analizar más a fondo, ya que tienen un volumen significativo de datos. Esto puede ayudar a identificar tendencias y patrones específicos de estas ciudades.")

# Distribución de calificaciones por ciudad (Top 10 ciudades con más reviews)
top_cities = city_counts.index[:10]
df_top_cities = df[df['city_x'].isin(top_cities)]

plt.figure(figsize=(14, 10))
sns.boxplot(data=df_top_cities, x='city_x', y='stars')
plt.title('Distribución de calificaciones por ciudad (Top 10)')
plt.xlabel('Ciudad')
plt.ylabel('Calificación')
plt.xticks(rotation=45)
plt.show()
print("Conclusión: Hay variabilidad en las calificaciones entre las diferentes ciudades. Por ejemplo, algunas ciudades muestran una mayor dispersión en las calificaciones, lo que podría indicar una mayor variabilidad en la calidad de los locales.")

# Categorías más comunes
plt.figure(figsize=(12, 8))
category_counts = df['categories_x'].value_counts().head(20)
sns.barplot(y=category_counts.index, x=category_counts.values, palette='viridis')
plt.title('Top 20 categorías más comunes')
plt.xlabel('Número de reviews')
plt.ylabel('Categoría')
plt.show()
print("Conclusión: Las categorías más comunes incluyen restaurantes, bares y cafeterías. Esto es útil para entender el contexto general de los locales en el dataset.")

# Análisis específico para 'Pizza Hut'
df_pizza_hut = df[df['name'].str.contains('Pizza Hut', case=False, na=False)]

# Información básica de Pizza Hut
print("\nInformación del dataframe de Pizza Hut:")
print(df_pizza_hut.info())

# Estadísticas descriptivas de Pizza Hut
print("\nEstadísticas descriptivas de Pizza Hut:")
print(df_pizza_hut.describe(include='all'))

# Distribución de las calificaciones para Pizza Hut
plt.figure(figsize=(10, 6))
sns.histplot(df_pizza_hut['stars'], bins=20, kde=True)
plt.title('Distribución de las calificaciones de Pizza Hut')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: Las calificaciones de Pizza Hut tienden a estar centradas en 1 y 2 estrellas, similar a la distribución general. Esto sugiere que los clientes tienen una percepción moderadamente positiva de los locales de Pizza Hut.")

# Distribución de la polaridad para Pizza Hut
plt.figure(figsize=(10, 6))
sns.histplot(df_pizza_hut['polarity'], bins=20, kde=True)
plt.title('Distribución de la polaridad de Pizza Hut')
plt.xlabel('Polaridad')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: La polaridad de las reseñas de Pizza Hut muestra una tendencia negativa, lo que indica que las opiniones son generalmente desfavorables.")

# Distribución de la subjetividad para Pizza Hut
plt.figure(figsize=(10, 6))
sns.histplot(df_pizza_hut['subjectivity'], bins=20, kde=True)
plt.title('Distribución de la subjetividad de Pizza Hut')
plt.xlabel('Subjetividad')
plt.ylabel('Frecuencia')
plt.show()
print("Conclusión: Las reseñas de Pizza Hut tienen una distribución de subjetividad desigual, lo que sugiere que los clientes son desequilibrados en términos de opiniones subjetivas y objetivas.")

# Conteo de reviews de Pizza Hut por ciudad
plt.figure(figsize=(12, 8))
pizza_hut_city_counts = df_pizza_hut['city_x'].value_counts().head(20)
sns.barplot(y=pizza_hut_city_counts.index, x=pizza_hut_city_counts.values, palette='viridis')
plt.title('Top 20 ciudades con más reviews de Pizza Hut')
plt.xlabel('Número de reviews')
plt.ylabel('Ciudad')
plt.show()
print("Conclusión: Las ciudades con más reviews de Pizza Hut pueden ser áreas prioritarias para realizar análisis de satisfacción y rendimiento, ya que estos lugares tienen una gran cantidad de feedback de los clientes.")

```
